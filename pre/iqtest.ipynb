```python
import tensorflow as tf
```

```python
train_x = [1,2,3,4,5,6,7]
train_y = [3,5,7,9,11,13,15]
```

```python
a = tf.Variable(0.1)
b = tf.Variable(0.2)

예측_y = train_x * a + b
```

```python
def lossfunc():
  예측_y = train_x * a + b
  return tf.keras.losses.mse(train_y, 예측_y) # mean squared error 평균[(예측1 - 실제1)^2 + (예측2 - 실제2)^2]
```

```python
opt = tf.keras.optimizers.legacy.Adam(learning_rate=0.01)

for i in range(10000):
  opt.minimize(lossfunc, var_list=[a,b]) # minimize(손실함수, varlist)

print(a.numpy(), b.numpy()) 
```
