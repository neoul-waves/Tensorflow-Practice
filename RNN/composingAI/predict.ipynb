```python
import tensorflow as tf
import numpy as np
```

```python
abc = open('./abcnotation/freedomdive.abc', 'r').read()

uniqueText = list(set(abc))
uniqueText.sort()
```

```python
textToNum = {}
numToText = {}

for i, data in enumerate(uniqueText):
  textToNum[data] = i
  numToText[i] = data
```

```python
# 텍스트 전부 숫자로 치환
숫자화 = []

for i in abc:
  숫자화.append(textToNum[i])

print(숫자화)
```

```python
model = tf.keras.models.load_model('../../models/RNN_Composing')
model.summary()
```

```python
첫입력값 = 숫자화[225: 225+25]
첫입력값 = tf.one_hot(첫입력값, 31)
첫입력값 = tf.expand_dims(첫입력값, axis=0)
```

```python
예측값 = model.predict(첫입력값)

predictMax = np.argmax(예측값[0])
print(numToText[predictMax])
print(numToText[숫자화[117+25]])
```

```python
music = []

for i in range(300):
  predict = model.predict(첫입력값)
  predict = np.argmax(predict[0])

  music.append(predict)
  next = 첫입력값.numpy()[0][1:]

  oneHotNum = tf.one_hot(predict, 31)

  첫입력값 = np.vstack([next, oneHotNum.numpy()])
  첫입력값 = tf.expand_dims(첫입력값, axis=0)
```

```python
abc = []

for i in music:
  abc.append(numToText[i])

print(''.join(abc))
```
